```python
import regex as rgx
def cleanmessage(text):
    removedots = rgx.compile(r'\.{2,}')
    removenewline = rgx.compile(r'(\r\n|\r|\n)')
    nodots = removedots.sub(' ', text.lower())
    nonewline = removenewline.sub(' ', nodots)
    return nonewline

newmsg = cleanmessage(df.message[1])
newmsg
```

```python
message = df.message.astype(str)
```

```python
newsrs = message.apply(cleanmessage)
```

```python
newsrs
```

```python
word_count=newsrs.apply(lambda txt: pd.value_counts(txt.split(" "))).sum(axis=0)
```

```python
word_count_sorted = word_count.sort_values(axis=0, ascending=False)
word_count_sorted
```

```python
d = {}
for row in converted.itertuples():
    if row[1] not in d:
        d[row[1]] = []
    d[row[1]].append(row[2])     
```
